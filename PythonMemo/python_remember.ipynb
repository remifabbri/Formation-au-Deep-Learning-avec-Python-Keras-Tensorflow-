# Exercice Produit Scalaire

```python
vecteur_u = [1.0, 2.0]
vecteur_v = [2.0, 2.0]

produitScalaire = 0
for i in range(len(vecteur_u)):
  produitScalaire += vecteur_u[i]*vecteur_v[i]

produitScalaire
```

# Tracés de fonctions

```python
import matplotlib.pyplot as plt
import math

x = [0.1*i for i in range(64)]
sin = [math.sin(i) for i in x]
cos = [math.cos(i) for i in x]
parabole = [i**2 for i in x]
# expo = [math.exp(i) for i in x]

plt.plot(x, sin, label='sin(x)', color='blue', linewidth=2.5)
plt.plot(x, cos, label='cos(x)', color='red', linewidth=0.5)
plt.plot(x, parabole, label='para(x)', color='green', linewidth=1)
# plt.plot(x, expo, label='expo(x)', color='pink', linewidth=2)
plt.ylabel('y')
plt.xlabel('x')
plt.legend(loc='upper left')
plt.show()
```
